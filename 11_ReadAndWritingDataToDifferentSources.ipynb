```python
import pandas as pd
```

```python
df=pd.read_csv(r"data\survey_results_public.csv")
df_schema=pd.read_csv(r"data\survey_results_schema.csv")
```

##### **`Now is the question how we can write data to csv file?`**

```python
filt=df["Country"]=="India"
india_df=df.loc[filt]
india_df.head()
```

##### **Lets export this new dataframe to a CSV data file:**

```python
india_df.to_csv(r"data\modified.csv")
```

##### To write my dataframe to tsv(Tab Seperated File)
- We use the to_csv method
- We exchange the extension of the new file to .tsv
- We specify the sep="\t"

```python
india_df.to_csv(r"data\modified.tsv",sep="\t")
```

##### **`Excel:`**
- To write to an old Excel format .xls => `pip install xlwt`
- To write to a new Excel format .xlsx => `pip install openyxl`
- To read an Excel file => `pip install xlrd`

```python
# Write to an excel file.
india_df.to_excel(r"data\modified.xlsx")
```

```python
# Read an excel file.
# I can directly specify my index column.
test=pd.read_excel(r"data\modified.xlsx",index_col="Respondent")
```

##### `JSON:`
- Write Data to JSON file
- Read Data from JSON file

```python
# Wrtiting to a JSON file
india_df.to_json(r"data\modified.json")
```

```python
# There is another way to write a JSON file
# What if I want my JSON file to be list like and noit dict like?
# This will read every reponse as a line
india_df.to_json(r"data\modified.json",orient="records",lines=True)
# Using this format we can write each row of the DataFrame represented as a separate JSON object,
# and each object written on a separate line
```

```python
# How to read JSON file:
test=pd.read_json(r"data\modified.json",orient="records",lines=True)
```

```python
test.head()
```

##### **`Connecting to database:`**

```python
from sqlalchemy import create_engine
import psycopg2
```

```python
# create an engine 
engine=create_engine("postgresql://postgres:admin@localhost:5432/mydb")
```

```python
# Write to a database
india_df.to_sql("sosoaldalouaa",engine)
```

```python
# To write to an exist table and I want to replace
india_df.to_sql("sosoaldalouaa",engine,if_exists="replace")
```

```python
# Read data from a database
sql_df=pd.read_sql("sosoaldalouaa",engine,index_col="Respondent")
```

```python
sql_df.head()
```

```python
# We can use read_sql_query to write a query
sql_df=pd.read_sql_query("SELECT * FROM sosoaldalouaa",engine,index_col="Respondent")
```
