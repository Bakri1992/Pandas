```python
import pandas as pd
```

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
df1=pd.DataFrame(people)
```

```python
df1
```

##### How to set and reset an index in pandas?
- To set a new index we use the method set_index("index_name")
- To reset the index to default we use reset_index()
- To keep the changes in both we set the arg inplace= True
- We can specify by reading or creating the dataframe using the index_col="col_name" arg

```python
df1.set_index("email",inplace=True)
```

```python
df1
```

```python
# We can access the index using df.index
df1.index
```

```python
# Now we can access the rows using the email
df1.loc["JaneDoe@email.com"]
```

```python
# We can as previous access a specific col in the dataframe
df1.loc["JaneDoe@email.com","last"]
```

```python
# Reset index
df1.reset_index(inplace=True)
```
